# 1. Employee salary Calculator
Question:
create a function that calculates the final salary of an employee after adding bonus and deducting tax.

### Question: Employee Salary Calculator
Create a function that calculates the final salary of an employee after adding bonus and deducting tax.

```python
def calculate_salary(basic_salary, bonus, tax):
    final_salary = basic_salary + bonus - tax
    return final_salary

basic_salary = float(input("Enter Basic Salary: ₹"))
bonus = float(input("Enter Bonus Amount: ₹"))
tax = float(input("Enter Tax Amount: ₹"))

final_salary = calculate_salary(basic_salary, bonus, tax)

print("\n----- Salary Details -----")
print("Basic Salary : ₹", basic_salary)
print("Bonus        : ₹", bonus)
print("Tax Deducted : ₹", tax)
print("--------------------------")
print("Final Salary : ₹", final_salary)
```

# 2. Student Result Analysis system

Question:

Create a function that accepts marks of multiple subjects using *args and calculate total, average, and grade.

```python
def student_result(*marks):
    total = sum(marks)
    average = total / len(marks)

    if average >= 90:
        grade = "A+"
    elif average >= 80:
        grade = "A"
    elif average >= 70:
        grade = "B"
    elif average >= 60:
        grade = "C"
    elif average >= 50:
        grade = "D"
    else:
        grade = "F"

    print("\n----- Student Result -----")
    print("Total Marks :", total)
    print("Average     :", average)
    print("Grade       :", grade)

n = int(input("Enter number of subjects: "))

marks = []
for i in range(n):
    mark = float(input(f"Enter marks of Subject {i+1}: "))
    marks.append(mark)

student_result(*marks)9
```

# Customer information Management System

Question:

use **kwargs to store customer details dynamically

```python
def customer_info(**details):
    print("\n- Customer Details -")
    for key, value in details.items():
        print(f"{key} : {value}")

name = input("Enter Customer Name: ")
age = input("Enter Age: ")
city = input("Enter City: ")
phone = input("Enter Phone Number: ")

customer_info(
    Name=name,
    Age=age,
    City=city,
    Phone=phone
)
```

# make a calculator


```python
def calculator():
    print("Simple Calculator")

    def get_number(prompt):
        while True:
            try:
                num_str = input(prompt)
                return float(num_str)
            except ValueError:
                print("Invalid input. Please enter a valid number.")

    num1 = get_number("Enter first number: ")
    operator = input("Enter operator (+, -, *, /): ")
    num2 = get_number("Enter second number: ")

    if operator == '+':
        result = num1 + num2
    elif operator == '-':
        result = num1 - num2
    elif operator == '*':
        result = num1 * num2
    elif operator == '/':
        if num2 != 0:
            result = num1 / num2
        else:
            print("Error: Division by zero is not allowed.")
            return
    else:
        print("Error: Invalid operator.")
        return

    print(f"Result: {num1} {operator} {num2} = {result}")

calculator()
```
