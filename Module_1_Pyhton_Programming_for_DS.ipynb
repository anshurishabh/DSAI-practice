3.Create a simple login validation system using a username and password

```python
CORRECT_USERNAME = "anshu"
CORRECT_PASSWORD = "password123"
attempts = 3

print("--- Welcome to the Secure Login System ---")
while attempts > 0:
    input_username = input("Enter Username: ")
    input_password = input("Enter Password: ")
    if input_username == CORRECT_USERNAME and input_password == CORRECT_PASSWORD:
        print("\nLogin Successful! Welcome back.")
        break
    else:
        attempts -= 1
        print(f"\nInvalid username or password. Remaining attempts: {attempts}")
        print("-" * 30)
if attempts == 0:
    print("\nAccount Locked. Too many failed attempts.")
```

**3.If-Elif-Else-Statement**

it is used to check the multiple conditions in a program . it execute a block of code when its condition evaluates to True after previous conditions evaluate to False

**4.Nasted If- Else Statemant**

it is an if-else statement placed inside another if or else block . it is used to check conditions within another condition

```python
age = 70
is_member = True
if age>= 60:
  if is_member:
    print("30% senior discount!")
  else:
    print("20% senior discount!")
else:
  print("Not eligible for the senior discount!")
```

# Loops

**1.For Loop**

A for loop over a sequence

```python
for i in range (1,6):
  print(i)
```

Example 2. Iterate through a List

```python
fruits = ["Apple", "Banana", "Mango"]
for fruit in fruits:
  print(fruit)
```

**2. While Loop**

A while loop runs as long as condition is True

```python
count = 1
while count <= 500:
  print(count)
  count+=1
```

**3. Loop control Statement**

A. break
   
   stops the loop immediately

```python
for i in range (1,10):
  if i==5:
    break
  print(i)
```

**B. continue**

Skips the current iteration

```python
for i in range (1,6):
  if i == 3:
    continue
  print(i)
```

**C. Pass**

placeholder statement

```python
for i in range (5):
  pass
print("Loop Completed")
```
